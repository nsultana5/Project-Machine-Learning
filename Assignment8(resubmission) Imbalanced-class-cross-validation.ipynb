# Task:2 #SMOTETomek Sampling

```python
#task2:Do SMOTETomek sampling on given dataset and create ML model and evaluate model by cross validation
```

```python
from imblearn.combine import SMOTETomek
```

```python
sm=SMOTETomek(random_state=42)   #craeting an object
```

```python
y.value_counts()  #before sampling
```

```python
X_sm,Y_sm=sm.fit_resample(x,y)
```

```python
X_sm.shape
```

```python
Y_sm.shape
```

```python
Y_sm.value_counts()
```

```python
X_sm.head()
```

# Task 3 : Near Miss for udner sampling

```python
#task 3:Do undersampling on given dataset and create Ml model and evaluate model by cross validation 
```

```python
from imblearn.under_sampling import NearMiss
```

```python
nm=NearMiss()  #creating an object
```

```python
X_nm,Y_nm=nm.fit_resample(x,y)
```

```python
X_nm.shape
```

```python
Y_nm.shape
```

```python
X_nm.head()
```

```python
Y_nm.value_counts()
```
